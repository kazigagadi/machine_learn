```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
plt.rcParams['font.sans-serif']=['SimHei']# SimSun Microsoft YaHei YouYuan FangSong
plt.rcParams['font.size']=12# SimSun Microsoft YaHei YouYuan FangSong
plt.rcParams['axes.unicode_minus']=False
```

```python
plt.plot([-1,2,3,4,5],[3,-1,4,5,2])
plt.ylabel("y轴")
plt.xlabel("x轴")
# plt.xlim(-5,10)# 设置x轴坐标范围
# plt.ylim(-2,8)# 设置y轴坐标范围
plt.axis([-2,8,-3,6])
# plt.savefig(r"D:\cs2433_文件迁移\桌面\excel&csv\折线图.jpg",dpi=600)# 保存图片
```

```python
# 更多折线图
a=np.linspace(0,10,10)
plt.plot(a,np.sin(a),color='#DC143C',linestyle='-.')# red blue yellow,# linewidth=4
plt.plot(a,np.cos(a),color='#D15FEE',linestyle='--')
# plt.plot(a+1,np.cos(a),color='y',linestyle=':')
# plt.plot(a,np.sin(a),'r-.',a,np.cos(a),'b--',a+1,np.sin(a),'ko')
```

```python
# 更多折线图
a=np.linspace(0,10,10)
plt.plot(a,a,'rx',a,a+1,'b+',a,a+2,'bd')
```

```python
# 更多折线图
plt.figure(figsize=(12,8))
a=np.linspace(0,10,100)
plt.plot(a,np.tan(a))
plt.xlabel("tan(x) x轴",fontproperties="FangSong",fontsize=20,color='b')
plt.ylabel("tan(x) y轴",fontproperties="FangSong",fontsize=20,color='r')
plt.title("Matplotlib绘制简易折线图",fontproperties="FangSong",fontsize=20,color='g')
```

```python
# 添加text文本
a=np.linspace(0,10,100)
plt.plot(a,np.sin(a),color='#DC143C',linestyle='-.')# red blue yellow,# linewidth=4
plt.plot(a,np.cos(a),color='#D15FEE',linestyle='--')
plt.text(4,0.5,"文本1",fontsize=16,color='b')
plt.text(7,0,"文本2",fontsize=16,color='r')
# plt.grid(True)
```

```python
plt.figure(figsize=(16,10))# 画板大小
a=np.arange(0,5,0.02)
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.xlabel('横轴：时间',fontproperties='SimHei',fontsize=20,color='b')
plt.ylabel('纵轴：振幅',fontproperties='SimHei',fontsize=20,color='r')
plt.title('时间与振幅的波动图')
plt.text(3,0,'文本信息1',fontsize=30)
plt.text(2,-1,'文本信息2',fontsize=30)
plt.annotate('小箭头',xytext=(1,-0.25),xy=(3,0),arrowprops=dict(shrink=0.1,width=4,color='b'))
plt.grid(True)
```

```python
# 多子图的绘图风格
plt.figure(figsize=(12,6))
a=np.linspace(0,10,100)
plt.subplot(2,1,1)
plt.plot(a,np.sin(a),'r--')
plt.subplot(2,1,2)
plt.plot(a,np.sin(a),'b:')
```

```python
plt.figure(figsize=(12,12))
color='rybg'
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.plot(a,np.sin(a),color=color[i])
    plt.text(5,0,'文本'+str(i+1),ha='center')
```

```python
# 面向对象的绘图风格
fig,ax=plt.subplots(2,1)
ax[0].plot(a,np.sin(a),'r--')
ax[1].plot(a,np.cos(a),'b-.')
```

```python
# 面向对象的绘图风格
fig,ax=plt.subplots(2,2,figsize=(12,12))
ax[0,0].plot(a,np.sin(a),'r-')
ax[0,1].plot(a,np.sin(a),'g:')
ax[1,0].plot(a,np.cos(a),'r-.')
ax[1,1].plot(a,np.cos(a),'b--')
```
