```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_20newsgroups #新闻数据集
from sklearn.model_selection import train_test_split #划分数据集
from sklearn.feature_extraction.text import TfidfVectorizer #特征抽取TF-IDF
from sklearn.naive_bayes import MultinomialNB #贝叶斯
from sklearn.metrics import classification_report #精准率，召回率，F1评分
```

```python
#加载数据集
news=fetch_20newsgroups(subset='all')
#选取特征值x 目标值y
x=news.data
y=news.target
```

```python
#划分测试集和训练集
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#特征工程：特征值化
tf=TfidfVectorizer()
x_train=tf.fit_transform(x_train)
x_test=tf.transform(x_test)
```

```python
#使用 朴素贝叶斯进行建模
mu=MultinomialNB(alpha=1.0)
mu.fit(x_train,y_train)
```

```python
#预测
y_predict=mu.predict(x_test)
y_predict,y_test
```

```python
#准确率
mu.score(x_test,y_test)
```

```python
#f1评估
print(classification_report(y_test,y_predict,target_names=news.target_names))
```

```python
from sklearn.metrics import confusion_matrix #生产混淆矩阵
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns  #画图
```

```python
#设置中文字体
matplotlib.rcParams['font.family']='SimHei'
matplotlib.rcParams['font.size']=10
```

```python
mat=confusion_matrix(y_test,y_predict) #生成混淆矩阵
sns.heatmap(mat,square=True,annot=True,cbar=True) #annot=True加载标签，cbar=False不显示颜色标记
plt.xlabel('预测值')
plt.ylabel('真实值')
plt.show()
```
