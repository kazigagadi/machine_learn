```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
plt.rcParams['font.sans-serif']=['Simhei']
plt.rcParams['axes.unicode_minus']=False
```

```python
#直方图
ser=pd.Series([-1,2,4,3,5],index=list("abcde"))
ser.plot()
```

```python
titan=pd.read_csv(r"F:\DataAnalysis\python\资料\全日制3班_20200507总结\数据分析课件\titanic_train.csv")
# print(titan.index())
plt.style.use("ggplot")
age=titan["Age"]
plt.hist(age,bins=20,color="b",edgecolor="k",label="直方图",alpha=0.5)
plt.legend()
```

```python
plt.hist(age,bins=20,color="b",edgecolor="k",label="直方图",alpha=0.5
         ,normed=True,cumulative=True)
plt.legend()
```

```python
age_female=titan[titan['Sex']=="female"]['Age']
age_male=titan[titan['Sex']=="male"]['Age']
plt.hist(age_female,bins=20,color="b",edgecolor="k",label="feMale",alpha=0.5)
plt.hist(age_male,bins=20,color="r",edgecolor="k",label="Male",alpha=0.1)
plt.legend()
```
