```python
# 读取数据
chengjiao=pd.read_excel(r"C:\Users\ibf\Desktop\python数据分析课件整理\数据可视化\数据\区域进客来源.xlsx",sheet_name="成交")
jinke=pd.read_excel(r"C:\Users\ibf\Desktop\python数据分析课件整理\数据可视化\数据\区域进客来源.xlsx",sheet_name="进客")
```

```python
# print(chengjiao.head(5))
# print(jinke.head(5))
# 城市订单占比分析,杭州订单占比最高
ddzb=chengjiao.groupby("城市")["城市"].count()
print(ddzb/ddzb.sum())# 求占比数据
chengjiao.groupby("城市")["城市"].count().plot(kind="pie")
plt.axis("equal")
```

```python
# 筛选出上海的新签有效订单
index1=(chengjiao["城市"]=="上海市") & (chengjiao["合同类型"]=="新签") & (chengjiao["出租合同状态"]=="有效")
shcj=chengjiao.loc[index1,:]
shcj.groupby("城区")["城市"].count().plot(kind="bar",figsize=(10,8))
# shcj.groupby("城区")["业绩"].mean().plot(kind="bar",figsize=(10,8))
plt.title("上海市新签订单区域比较")
```

```python
from example.commons import Faker
from pyecharts import options as opts
from pyecharts.charts import Bar
```

```python
# 基础绘图
def bar_base() -> Bar:
    c = (
        Bar()
        .add_xaxis(list(gp.count().index))
        .add_yaxis("区域订单量",list(map(lambda x:int(x),gp.count().values)))
        .add_yaxis("区域业绩总和",list(np.round(gp1.sum().values)/1000))
        .set_global_opts(title_opts=opts.TitleOpts(title="Bar-基本示例", subtitle="我是副标题"),
            toolbox_opts=opts.ToolboxOpts())# 显示工具栏
    )
    return c
```

```python
bar_base().render()
```

```python
# 转化率=成交/进客
zhl=shcj.groupby("城区")["城市"].count()/shjk.groupby("城区")["城市"].count()
# print(zhl.fillna(0).map(lambda x:"%.1f%%"%(x*100)))
print(zhl.fillna(0).sort_values(ascending=False).plot(kind="bar",figsize=(10,8),color="red",alpha=0.5))
```

```python
# 上海市渠道进客情况
index3=(jinke["城市"]=="上海市")
qdjk=jinke.loc[index3,:]
qdjkgp=qdjk.groupby("客户来源")["城市"].count().sort_values()
qdjkgp.plot(kind="bar",figsize=(10,8),color="red",alpha=0.5)
```

```python
# 上海市渠道进客情况
index4=(chengjiao["城市"]=="上海市") & (chengjiao["合同类型"]=="新签") & (chengjiao["出租合同状态"]=="有效")
qdcj=chengjiao.loc[index4,:]
qdcjgp=qdcj.groupby("租客来源")["城市"].count().sort_values()
qdcjgp.plot(kind="bar",figsize=(10,8),color="red",alpha=0.5)
```

```python
left=qdcjgp.reset_index()# 成交的数据
right=qdjkgp.reset_index()# 进客的数据
left.columns=["客户来源","成交数量"]
right.columns=["客户来源","进客数量"]
data_merge=pd.merge(left,right,on="客户来源")
data_merge=data_merge.loc[data_merge["成交数量"]>3,:]
data_merge
```

```python
print(len(data_merge))
```

```python
plt.figure(figsize=(12,8))
bar_width = 0.45
# 绘图
plt.bar(np.arange(len(data_merge)),data_merge['成交数量'], label = '成交数量', color = 'steelblue', alpha = 0.5, width = bar_width)
plt.bar(np.arange(len(data_merge))+bar_width,data_merge['进客数量'], label = '进客数量', color = 'indianred', alpha = 0.5, width = bar_width)
# 添加轴标签
plt.xlabel('上海渠道来源')
plt.ylabel('数量')
# 添加标题
plt.title('上海区域渠道进客成交情况')
# 添加刻度标签
plt.xticks(np.arange(len(data_merge))+bar_width/2,data_merge['客户来源'])

# # 为每个条形图添加数值标签
for x,y in enumerate(data_merge['成交数量']):
    plt.text(x,y+5,'%s'%round(y),ha='center')

for x,y in enumerate(data_merge['进客数量']):
    plt.text(x+bar_width,y+5,'%s'%round(y),ha='center')
# 显示图例
plt.legend(loc='upper center')
```
