```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
plt.rcParams['font.sans-serif']=['Simhei']
plt.rcParams['axes.unicode_minus']=False
```

```python
#简易折线图
plt.plot([3,1,4,5,2])
plt.ylabel("y轴")
plt.xlabel("x轴")
# plt.ylim(-2,8)
# plt.xlim(-5,10)
plt.axis([-2,8,-3,6])
plt.savefig(r"F:\DataAnalysis\python\jpg\折线图.jpg")
plt.show()
```

```python
#更多折线图
a=np.linspace(0,10,100)
# plt.plot(a,np.sin(a),color='r',linestyle='-.')
# plt.plot(a,np.cos(a),color='b',linestyle='--')
# plt.plot(a+1,np.cos(a),color='y',linestyle=':')
plt.plot(a+1,np.cos(a),'r-.',a,np.sin(a),'b--',a+1,np.sin(a),'y:')
```

```python
#更多折线图
a=np.linspace(0,10,100)
plt.plot(a,a,'rx',a,a+1,'bd')
```

```python
#更多折线图
plt.figure(figsize=(12,6))
a=np.linspace(0,10,100)
plt.plot(a,np.tan(a))
plt.xlabel("tan(x) x轴",fontproperties="FangSong")
plt.ylabel("tan(y) y轴")
plt.title("呵呵")
```

```python
#添加文本
plt.figure(figsize=(12,6))
a=np.linspace(0,10,100)
plt.plot(a,np.tan(a))
plt.xlabel("tan(x) x轴",fontproperties="FangSong")
plt.ylabel("tan(y) y轴")
plt.title("呵呵")
plt.text(4,0.5,"kankan",fontsize=16,color='r')
```

```python
plt.figure(figsize=(16,10))# 画板大小
a=np.arange(0,5,0.02)
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.xlabel('横轴：时间',fontproperties='SimHei',fontsize=20,color='b')
plt.ylabel('纵轴：振幅',fontproperties='SimHei',fontsize=20,color='r')
plt.title('时间与振幅的波动图')
plt.text(3,0,'文本信息1',fontsize=30)
plt.text(2,-1,'文本信息2',fontsize=30)
plt.annotate('小箭头',xytext=(1,-0.25),xy=(3,0),arrowprops=dict(shrink=0.1,width=4,color='b'))
plt.grid(True)
```

```python
#多子图风格
a=np.linspace(0,10,100)
plt.subplot(2,1,1)
plt.plot(a,np.sin(a),'r--')
plt.subplot(2,1,2)
plt.plot(a,np.cos(a),'r--')
```

```python
plt.figure(figsize=(12,12))
color='rybg'
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.plot(a,np.sin(a),color=color[i])
    plt.text(5,0,'文本'+str(i+1),ha='center')
```

```python
#面向对象绘图风格
fig,ax=plt.subplots(2,1)
ax[0].plot(a,np.sin(a),'r--')
ax[1].plot(a,np.cos(a),'b-.')
```

```python
#面向对象绘图风格
fig,ax=plt.subplots(2,2,figsize=(12,12))
ax[0,0].plot(a,np.sin(a),'r--')
ax[0,1].plot(a,np.sin(a),'b--')
ax[1,0].plot(a,np.cos(a),'r--')
ax[1,1].plot(a,np.cos(a),'g-.')
```
